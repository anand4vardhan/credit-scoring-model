```python
import pandas as pd
train_data=pd.read_csv("R_Module_Day_7.2_Credit_Risk_Train_data.csv")
test_data=pd.read_csv("R_Module_Day_8.2_Credit_Risk_Test_data.csv")
```

```python
import pandas as pd
```

```python
train_data['Source']='Train'
test_data['Source']='Test'
full_data=pd.concat([train_data,test_data],axis=0)
```

```python
import numpy as np

#full_data.Dependents.dtype
```

```python
full_data.Dependents=np.where(full_data.Dependents=='3+',3,full_data.Dependents).astype(float)
```

```python
full_data.isnull().sum()
```

```python
for i in range(len(full_data.columns)):
    print(full_data.columns[i])
    if(full_data.iloc[:,i].dtype==object):
        if(full_data.iloc[:,i].isnull().sum()>0):
            full_data.iloc[:,i].fillna(full_data.iloc[:,i][full_data.Source=='Train'].mode()[0], inplace=True)
            #print("1")
    else:
        #print("2")
        full_data.iloc[:,i].fillna(full_data.iloc[:,i][full_data.Source=='Train'].median(), inplace=True)
```

```python
list1=['Gender','Married','Education','Self_Employed','Property_Area']
```

```python
df_dummies=pd.get_dummies(full_data[list1],drop_first=True)
```

```python
full_data.Loan_Status=np.where(full_data.Loan_Status=='N',1,0)
```

```python
full_data1=full_data.drop(cols_to_rem, axis=1).copy()
```

```python
full_data2=pd.concat([full_data1,df_dummies],axis=1)
```

```python
full_data2.shape
```

```python
full_data2
```

```python
train=full_data2.loc[full_data2.Source=='Train',].drop('Source',axis=1).copy()
```

```python
test=full_data2.loc[full_data2.Source=='Test',].drop('Source',axis=1).copy()
```

```python
train.columns
```

```python
train_x=train.drop('Loan_Status',axis= 1).copy()
train_y=train['Loan_Status'].copy()
test_x=test.drop('Loan_Status',axis= 1).copy()
test_y=test['Loan_Status'].copy()
```

```python
#build a model on svm
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
M1=SVC()
```

```python
M1_model=M1.fit(train_x,train_y)
```

```python
Test_Pred=M1_model.predict(test_x)
```

```python
#Test_Pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_mat=confusion_matrix(test_y,Test_Pred)
```

```python
confusion_mat
```

```python
report=classification_report(test_y,Test_Pred)
```

```python
print(report)
```
